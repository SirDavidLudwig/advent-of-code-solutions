```python
import io
import itertools
import numpy as np
import os
import queue
import sys
```

```python
OP_CODE_MAP = {}

def operator(opcode, nparams, label):
    def decorate(opfunc):
        setattr(opfunc, "opcode", opcode)
        setattr(opfunc, "nparams", nparams)
        setattr(opfunc, "label", label)
        return opfunc
    return decorate

class PipeIO(io.TextIOBase):
    def __init__(self, newline='\n'):
        self.newline = newline
        self.buffer = ""

    def read(self):
        result = self.buffer
        self.buffer = ""
        return result

    def readline(self):
        try:
            index = self.buffer.index(self.newline) + 1
        except:
            index =len(self.buffer)
        result = self.buffer[:index]
        self.buffer = self.buffer[index:]
        return result

    def write(self, s):
        self.buffer += s

    def __next__(self):
        if len(self.buffer) == 0:
            raise StopIteration()
        return self.readline()

class IntcodeComputer:
    OPS = OP_CODE_MAP

    STATE_HALTED  = 0
    STATE_IDLE    = 1
    STATE_RUNNING = 2

    @staticmethod
    def program_from_string(program):
        return np.array(list(map(int, program.split(','))))

    def __init__(self, output_prefix=""):
        self.input_buffer = sys.stdin
        self.output_buffer = sys.stdout
        self.ops = self.build_operators()
        self.state = IntcodeComputer.STATE_HALTED

    def __getitem__(self, address):
        return self.memory[address]

    def __setitem__(self, address, value):
        self.memory[address] = value

    def build_operators(self):
        ops = filter(lambda key: hasattr(getattr(self, key), "opcode"), dir(self))
        return {getattr(self, op).opcode: getattr(self, op) for op in ops}

    def set_state(self, state):
        self.state = state

    def fetch(self, address, mode='1'):
        match mode:
            case '0': # position mode
                return self[self[address]]
            case '1': # immediate mode
                return self[address]
        raise Exception("Attempted to fetch with unknown mode:", mode)

    def store(self, address, value, mode='1'):
        match mode:
            case '0': # position mode
                self[self[address]] = value
            case '1': # immediate mode
                self[address] = value
            case _:
                raise Exception("Attempted to store with unknown mode:", mode)

    @operator(opcode=1, nparams=3, label="ADD")
    def op_add(self, a, b, dest, modes):
        self.store(self[dest], self.fetch(a, modes[0]) + self.fetch(b, modes[1]))
        self.instruction_pointer += 4

    @operator(opcode=2, nparams=3, label="MUL")
    def op_mul(self, a, b, dest, modes):
        self.store(self[dest], self.fetch(a, modes[0]) * self.fetch(b, modes[1]))
        self.instruction_pointer += 4

    @operator(opcode=3, nparams=1, label="READ")
    def op_read(self, dest, modes):
        try:
            value = int(next(self.input_buffer))
            self.store(self[dest], int(value))
            self.instruction_pointer += 2
        except (StopIteration, ValueError):
            self.set_state(IntcodeComputer.STATE_IDLE)

    @operator(opcode=4, nparams=1, label="WRITE")
    def op_write(self, address, modes):
        self.output_buffer.write(f"{self.fetch(address, modes[0])}\n")
        self.output_buffer.flush()
        self.instruction_pointer += 2

    @operator(opcode=5, nparams=2, label="JUMP_IF_TRUE")
    def op_jump_if_true(self, condition, instruction_address, modes):
        if self.fetch(condition, modes[0]) == 0:
            self.instruction_pointer += 3
            return
        self.instruction_pointer = self.fetch(instruction_address, modes[1])

    @operator(opcode=6, nparams=2, label="JUMP_IF_FALSE")
    def op_jump_if_false(self, condition, instruction_address, modes):
        if self.fetch(condition, modes[0]) != 0:
            self.instruction_pointer += 3
            return
        self.instruction_pointer = self.fetch(instruction_address, modes[1])

    @operator(opcode=7, nparams=3, label="LESS_THAN")
    def op_less_than(self, a, b, dest, modes):
        self.store(self[dest], int(self.fetch(a, modes[0]) < self.fetch(b, modes[1])))
        self.instruction_pointer += 4

    @operator(opcode=8, nparams=3, label="GREATER")
    def op_equals(self, a, b, dest, modes):
        self.store(self[dest], int(self.fetch(a, modes[0]) == self.fetch(b, modes[1])))
        self.instruction_pointer += 4

    @operator(opcode=99, nparams=0, label="HALT")
    def op_halt(self, modes):
        self.set_state(IntcodeComputer.STATE_HALTED)

    def step(self, debug=False):
        modes, opcode = self[self.instruction_pointer] // 100, self[self.instruction_pointer] % 100
        instruction = self.ops[opcode]
        modes = f"{str(modes):>03}"[:-instruction.nparams-1:-1]
        params = list(range(self.instruction_pointer + 1, self.instruction_pointer + instruction.nparams + 1))
        if debug:
            print("Instruction pointer:", self.instruction_pointer)
            print("Instruction:", opcode, f"({instruction.label})"),
            print("Parameter Modes:", modes)
            print("Parameter Values:", [self.fetch(p, m) for p, m in zip(params, modes)])
            print("Memory Dump:")
            print(self.memory)
            print()
        instruction(*params, modes)
        return self.state

    def execute(self, debug=False, num_steps=np.inf):
        self.set_state(IntcodeComputer.STATE_RUNNING)
        step_count = 0
        while self.state == IntcodeComputer.STATE_RUNNING:
            self.step(debug)
            if step_count > num_steps:
                self.set_state(IntcodeComputer.STATE_IDLE)
            step_count += 1
        return self.state

    def boot(self, program):
        self.memory = program.copy()
        self.instruction_pointer = 0
        self.state = IntcodeComputer.STATE_IDLE

    def pipe_to(self, other_computer):
        pipe = PipeIO()
        self.set_output_buffer(pipe)
        other_computer.set_input_buffer(pipe)

    def set_input_buffer(self, input_buffer):
        self.input_buffer = iter(input_buffer) if input_buffer else iter([])

    def set_output_buffer(self, output_buffer):
        self.output_buffer = output_buffer
```

<p style="border-color: rgb(204, 204, 204); color: rgb(204, 204, 204);">In feedback loop mode, the amplifiers need <em style="border-color: rgb(255, 255, 255); color: rgb(255, 255, 255); text-shadow: rgb(255, 255, 255) 0px 0px 5px;">totally different phase settings</em>: integers from <code style="border: 1px solid rgb(51, 51, 64); color: rgb(204, 204, 204); background: rgb(16, 16, 26);">5</code> to <code style="border: 1px solid rgb(51, 51, 64); color: rgb(204, 204, 204); background: rgb(16, 16, 26);">9</code>, again each used exactly once. <p style="border-color: rgb(204, 204, 204); color: rgb(204, 204, 204);">Don't restart the Amplifier Controller Software on any amplifier during this process. Each one should continue receiving and sending signals until it halts.</p>
<p style="border-color: rgb(204, 204, 204); color: rgb(204, 204, 204);">All signals sent or received in this process will be between pairs of amplifiers except the very first signal and the very last signal. When this happens, the last output signal from amplifier <code style="border: 1px solid rgb(51, 51, 64); color: rgb(204, 204, 204); background: rgb(16, 16, 26);">E</code> is sent to the thrusters.
