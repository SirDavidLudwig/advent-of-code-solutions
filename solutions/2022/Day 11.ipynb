```python
from collections import defaultdict
import numpy as np
import queue
import re
```

```python
class Monkey:
    def __init__(self, identifier, starting_items, operation, test, test_true, test_false):
        self.identifier = identifier
        self.starting_items = starting_items
        self.worry = self.build_worry(operation)
        self.test = test
        self.test_true = test_true
        self.test_false = test_false
        self.init()

    def build_worry(self, operation):
        _, op, right = operation
        op = int.__add__ if op == '+' else int.__mul__
        if right == "old":
            return lambda old: op(old, old)
        right = int(right)
        return lambda old: op(old, right)

    def init(self):
        self.num_items_inspected = 0
        self.q = queue.Queue()
        for item in self.starting_items:
            self.q.put(item)

    def catch_item(self, item):
        self.q.put(item)

    def throw_items(self, relief=lambda x: x // 3):
        thrown_items = []
        self.num_items_inspected += self.q.qsize()
        while not self.q.empty():
            item = self.q.get()
            item = relief(self.worry(item))
            to = self.test_true if item % self.test == 0 else self.test_false
            thrown_items.append((item, to))
        return thrown_items

    def __repr__(self):
        return str(self)

    def __str__(self):
        result = f"Monkey {self.identifier}\n"
        result += f"  Items: [{', '.join([str(item) for item in self.q.queue])}]"
        return result
```

```python
monkeys = []
with open("./puzzle_inputs/11.txt") as f:
    while f.readline():
        starting_items = list(map(int, re.findall(r"\d+", f.readline())))
        operation = re.findall(r"(old|\d+)\s(\+|\*)\s(old|\d+)", f.readline())[0]
        test = int(re.findall("\d+", f.readline())[0])
        test_true = int(re.findall("\d+", f.readline())[0])
        test_false = int(re.findall("\d+", f.readline())[0])
        monkeys.append(Monkey(len(monkeys), starting_items, operation, test, test_true, test_false))
        f.readline()
```

```python
for monkey in monkeys:
    monkey.init()

for r in range(20):
    for monkey in monkeys:
        for item, to in monkey.throw_items():
            # print(item)
            monkeys[to].catch_item(item)
```

```python
mod = int(np.product([monkey.test for monkey in monkeys]))
relief = lambda x: x % mod
```

```python
for monkey in monkeys:
    monkey.init()

for r in range(10000):
    for monkey in monkeys:
        for item, to in monkey.throw_items(relief=relief):
            monkeys[to].catch_item(item)
```
