```python
start_east = set()
start_south = set()
with open("./puzzle_inputs/25.txt") as f:
    for i, line in enumerate(f):
        for j, c in enumerate(line.strip()):
            if c == 'v':
                start_south.add((i,j))
            elif c == '>':
                start_east.add((i,j))
height, width = i+1, j+1
```

```python
height, width
```

```python
def step(east, south):
    changed = False
    east_step = set()
    for (i, j) in east:
        next_index = (i, (j+1) % width)
        if next_index in east or next_index in south:
            east_step.add((i,j))
        else:
            changed = True
            east_step.add(next_index)
    east = east_step

    south_step = set()
    for (i, j) in south:
        next_index = ((i+1) % height, j)
        if next_index in east or next_index in south:
            south_step.add((i,j))
        else:
            changed = True
            south_step.add(next_index)
    south = south_step
    return east, south, changed
```

```python
def draw(east, south):
    for i in range(height):
        for j in range(width):
            assert not ((i,j) in east and (i,j) in south)
            if (i,j) in east:
                print(">", end="")
            elif (i,j) in south:
                print("v", end="")
            else:
                print(".", end="")
        print()
```

```python
east = start_east.copy()
south = start_south.copy()

changed = True
step_count = 0
while changed:
    east, south, changed = step(east, south)
    step_count += 1
step_count
```
