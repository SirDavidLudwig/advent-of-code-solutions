<p style="border-color: rgb(204, 204, 204); color: rgb(204, 204, 204);">The Thermal Environment Supervision Terminal (TEST) starts by running a <em style="border-color: rgb(255, 255, 255); color: rgb(255, 255, 255); text-shadow: rgb(255, 255, 255) 0px 0px 5px;">diagnostic program</em> (your puzzle input).  The TEST diagnostic program will run on <a href="2" style="border-color: rgb(0, 153, 0); color: rgb(0, 153, 0);">your existing Intcode computer</a> after a few modifications:</p>
<p style="border-color: rgb(204, 204, 204); color: rgb(204, 204, 204);"><em style="border-color: rgb(255, 255, 255); color: rgb(255, 255, 255); text-shadow: rgb(255, 255, 255) 0px 0px 5px;">First</em>, you'll need to add <em style="border-color: rgb(255, 255, 255); color: rgb(255, 255, 255); text-shadow: rgb(255, 255, 255) 0px 0px 5px;">two new instructions</em>:</p>
<ul style="border-color: rgb(204, 204, 204); color: rgb(204, 204, 204);">
<li style="border-color: rgb(204, 204, 204); color: rgb(204, 204, 204);">Opcode <code style="border: 1px solid rgb(51, 51, 64); color: rgb(204, 204, 204); background: rgb(16, 16, 26);">3</code> takes a single integer as <em style="border-color: rgb(255, 255, 255); color: rgb(255, 255, 255); text-shadow: rgb(255, 255, 255) 0px 0px 5px;">input</em> and saves it to the position given by its only parameter. For example, the instruction <code style="border: 1px solid rgb(51, 51, 64); color: rgb(204, 204, 204); background: rgb(16, 16, 26);">3,50</code> would take an input value and store it at address <code style="border: 1px solid rgb(51, 51, 64); color: rgb(204, 204, 204); background: rgb(16, 16, 26);">50</code>.</li>
<li style="border-color: rgb(204, 204, 204); color: rgb(204, 204, 204);">Opcode <code style="border: 1px solid rgb(51, 51, 64); color: rgb(204, 204, 204); background: rgb(16, 16, 26);">4</code> <em style="border-color: rgb(255, 255, 255); color: rgb(255, 255, 255); text-shadow: rgb(255, 255, 255) 0px 0px 5px;">outputs</em> the value of its only parameter. For example, the instruction <code style="border: 1px solid rgb(51, 51, 64); color: rgb(204, 204, 204); background: rgb(16, 16, 26);">4,50</code> would output the value at address <code style="border: 1px solid rgb(51, 51, 64); color: rgb(204, 204, 204); background: rgb(16, 16, 26);">50</code>.</li>
</ul>
<p style="border-color: rgb(204, 204, 204); color: rgb(204, 204, 204);">Programs that use these instructions will come with documentation that explains what should be connected to the input and output. The program <code style="border: 1px solid rgb(51, 51, 64); color: rgb(204, 204, 204); background: rgb(16, 16, 26);">3,0,4,0,99</code> outputs whatever it gets as input, then halts.</p>
<p style="border-color: rgb(204, 204, 204); color: rgb(204, 204, 204);"><em style="border-color: rgb(255, 255, 255); color: rgb(255, 255, 255); text-shadow: rgb(255, 255, 255) 0px 0px 5px;">Second</em>, you'll need to add support for <em style="border-color: rgb(255, 255, 255); color: rgb(255, 255, 255); text-shadow: rgb(255, 255, 255) 0px 0px 5px;">parameter modes</em>:</p>
<p style="border-color: rgb(204, 204, 204); color: rgb(204, 204, 204);">Each parameter of an instruction is handled based on its parameter mode.  Right now, your ship computer already understands parameter mode <code style="border: 1px solid rgb(51, 51, 64); color: rgb(204, 204, 204); background: rgb(16, 16, 26);">0</code>, <em style="border-color: rgb(255, 255, 255); color: rgb(255, 255, 255); text-shadow: rgb(255, 255, 255) 0px 0px 5px;">position mode</em>, which causes the parameter to be interpreted as a <em style="border-color: rgb(255, 255, 255); color: rgb(255, 255, 255); text-shadow: rgb(255, 255, 255) 0px 0px 5px;">position</em> - if the parameter is <code style="border: 1px solid rgb(51, 51, 64); color: rgb(204, 204, 204); background: rgb(16, 16, 26);">50</code>, its value is <em style="border-color: rgb(255, 255, 255); color: rgb(255, 255, 255); text-shadow: rgb(255, 255, 255) 0px 0px 5px;">the value stored at address <code style="border: 1px solid rgb(51, 51, 64); color: rgb(255, 255, 255); text-shadow: rgb(255, 255, 255) 0px 0px 5px; background: rgb(16, 16, 26);">50</code> in memory</em>. Until now, all parameters have been in position mode.</p>
<p style="border-color: rgb(204, 204, 204); color: rgb(204, 204, 204);">Now, your ship computer will also need to handle parameters in mode <code style="border: 1px solid rgb(51, 51, 64); color: rgb(204, 204, 204); background: rgb(16, 16, 26);">1</code>, <em style="border-color: rgb(255, 255, 255); color: rgb(255, 255, 255); text-shadow: rgb(255, 255, 255) 0px 0px 5px;">immediate mode</em>. In immediate mode, a parameter is interpreted as a <em style="border-color: rgb(255, 255, 255); color: rgb(255, 255, 255); text-shadow: rgb(255, 255, 255) 0px 0px 5px;">value</em> - if the parameter is <code style="border: 1px solid rgb(51, 51, 64); color: rgb(204, 204, 204); background: rgb(16, 16, 26);">50</code>, its value is simply <em style="border-color: rgb(255, 255, 255); color: rgb(255, 255, 255); text-shadow: rgb(255, 255, 255) 0px 0px 5px;"><code style="border: 1px solid rgb(51, 51, 64); color: rgb(255, 255, 255); text-shadow: rgb(255, 255, 255) 0px 0px 5px; background: rgb(16, 16, 26);">50</code></em>.</p>
<p style="border-color: rgb(204, 204, 204); color: rgb(204, 204, 204);">Parameter modes are stored in the same value as the instruction's opcode.  The opcode is a two-digit number based only on the ones and tens digit of the value, that is, the opcode is the rightmost two digits of the first value in an instruction. Parameter modes are single digits, one per parameter, read right-to-left from the opcode: the first parameter's mode is in the hundreds digit, the second parameter's mode is in the thousands digit, the third parameter's mode is in the ten-thousands digit, and so on. Any missing modes are <code style="border: 1px solid rgb(51, 51, 64); color: rgb(204, 204, 204); background: rgb(16, 16, 26);">0</code>.</p>
<p style="border-color: rgb(204, 204, 204); color: rgb(204, 204, 204);">For example, consider the program <code style="border: 1px solid rgb(51, 51, 64); color: rgb(204, 204, 204); background: rgb(16, 16, 26);">1002,4,3,4,33</code>.</p>
<p style="border-color: rgb(204, 204, 204); color: rgb(204, 204, 204);">The first instruction, <code style="border: 1px solid rgb(51, 51, 64); color: rgb(204, 204, 204); background: rgb(16, 16, 26);">1002,4,3,4</code>, is a <em style="border-color: rgb(255, 255, 255); color: rgb(255, 255, 255); text-shadow: rgb(255, 255, 255) 0px 0px 5px;">multiply</em> instruction - the rightmost two digits of the first value, <code style="border: 1px solid rgb(51, 51, 64); color: rgb(204, 204, 204); background: rgb(16, 16, 26);">02</code>, indicate opcode <code style="border: 1px solid rgb(51, 51, 64); color: rgb(204, 204, 204); background: rgb(16, 16, 26);">2</code>, multiplication.  Then, going right to left, the parameter modes are <code style="border: 1px solid rgb(51, 51, 64); color: rgb(204, 204, 204); background: rgb(16, 16, 26);">0</code> (hundreds digit), <code style="border: 1px solid rgb(51, 51, 64); color: rgb(204, 204, 204); background: rgb(16, 16, 26);">1</code> (thousands digit), and <code style="border: 1px solid rgb(51, 51, 64); color: rgb(204, 204, 204); background: rgb(16, 16, 26);">0</code> (ten-thousands digit, not present and therefore zero):</p>
<pre style="border: 1px solid rgb(51, 51, 64); color: rgb(204, 204, 204); display: inline-block;"><code style="border-color: rgb(204, 204, 204); color: rgb(204, 204, 204);">ABCDE
 1002

DE - two-digit opcode,      02 == opcode 2
 C - mode of 1st parameter,  0 == position mode
 B - mode of 2nd parameter,  1 == immediate mode
 A - mode of 3rd parameter,  0 == position mode,
                                  omitted due to being a leading zero
</code></pre>
<p style="border-color: rgb(204, 204, 204); color: rgb(204, 204, 204);">This instruction multiplies its first two parameters.  The first parameter, <code style="border: 1px solid rgb(51, 51, 64); color: rgb(204, 204, 204); background: rgb(16, 16, 26);">4</code> in position mode, works like it did before - its value is the value stored at address <code style="border: 1px solid rgb(51, 51, 64); color: rgb(204, 204, 204); background: rgb(16, 16, 26);">4</code> (<code style="border: 1px solid rgb(51, 51, 64); color: rgb(204, 204, 204); background: rgb(16, 16, 26);">33</code>). The second parameter, <code style="border: 1px solid rgb(51, 51, 64); color: rgb(204, 204, 204); background: rgb(16, 16, 26);">3</code> in immediate mode, simply has value <code style="border: 1px solid rgb(51, 51, 64); color: rgb(204, 204, 204); background: rgb(16, 16, 26);">3</code>. The result of this operation, <code style="border: 1px solid rgb(51, 51, 64); color: rgb(204, 204, 204); background: rgb(16, 16, 26);">33 * 3 = 99</code>, is written according to the third parameter, <code style="border: 1px solid rgb(51, 51, 64); color: rgb(204, 204, 204); background: rgb(16, 16, 26);">4</code> in position mode, which also works like it did before - <code style="border: 1px solid rgb(51, 51, 64); color: rgb(204, 204, 204); background: rgb(16, 16, 26);">99</code> is written to address <code style="border: 1px solid rgb(51, 51, 64); color: rgb(204, 204, 204); background: rgb(16, 16, 26);">4</code>.</p>
<p style="border-color: rgb(204, 204, 204); color: rgb(204, 204, 204);">Parameters that an instruction writes to will <em style="border-color: rgb(255, 255, 255); color: rgb(255, 255, 255); text-shadow: rgb(255, 255, 255) 0px 0px 5px;">never be in immediate mode</em>.</p>
<p style="border-color: rgb(204, 204, 204); color: rgb(204, 204, 204);"><em style="border-color: rgb(255, 255, 255); color: rgb(255, 255, 255); text-shadow: rgb(255, 255, 255) 0px 0px 5px;">Finally</em>, some notes:</p>
<ul style="border-color: rgb(204, 204, 204); color: rgb(204, 204, 204);">
<li style="border-color: rgb(204, 204, 204); color: rgb(204, 204, 204);">It is important to remember that the instruction pointer should increase by <em style="border-color: rgb(255, 255, 255); color: rgb(255, 255, 255); text-shadow: rgb(255, 255, 255) 0px 0px 5px;">the number of values in the instruction</em> after the instruction finishes. <li style="border-color: rgb(204, 204, 204); color: rgb(204, 204, 204);">Integers can be negative: <code style="border: 1px solid rgb(51, 51, 64); color: rgb(204, 204, 204); background: rgb(16, 16, 26);">1101,100,-1,4,0</code> is a valid program (find <code style="border: 1px solid rgb(51, 51, 64); color: rgb(204, 204, 204); background: rgb(16, 16, 26);">100 + -1</code>, store the result in position <code style="border: 1px solid rgb(51, 51, 64); color: rgb(204, 204, 204); background: rgb(16, 16, 26);">4</code>).</li>

```python
def static_vars(**kwargs):
    def decorate(func):
        for k in kwargs:
            setattr(func, k, kwargs[k])
        return func
    return decorate
```

```python
class IntcodeComputer:
    ADD   = 1
    MUL   = 2
    READ  = 3
    WRITE = 4
    HALT  = 99

    def __init__(self, output_prefix=""):
        self.output_prefix = output_prefix
        self.instructions = {
            IntcodeComputer.ADD: self.add,
            IntcodeComputer.MUL: self.mul,
            IntcodeComputer.READ: self.read,
            IntcodeComputer.WRITE: self.write
        }

    def __getitem__(self, address):
        return self.program[address]

    def __setitem__(self, address, value):
        self.program[address] = value

    def fetch(self, address, mode='1'):
        match mode:
            case '0': # position mode
                return self[self[address]]
            case '1': # immediate mode
                return self[address]
        raise Exception("Attempted to fetch with unknown mode:", mode)

    def store(self, address, value, mode='1'):
        match mode:
            case '0': # position mode
                self[self[address]] = value
            case '1': # immediate mode
                self[address] = value
            case _:
                raise Exception("Attempted to store with unknown mode:", mode)

    @static_vars(nparams=3)
    def add(self, a, b, dest, modes):
        self.store(self[dest], self.fetch(a, modes[0]) + self.fetch(b, modes[1]))

    @static_vars(nparams=3)
    def mul(self, a, b, dest, modes):
        self.store(self[dest], self.fetch(a, modes[0]) * self.fetch(b, modes[1]))

    @static_vars(nparams=1)
    def read(self, dest, _):
        self.store(self[dest], next(self.inputs))

    @static_vars(nparams=1)
    def write(self, address, modes):
        print(f"{self.output_prefix}{self.fetch(address, modes[0])}")

    def interp(self, i):
        modes, opcode = self[i] // 100, self[i] % 100

        instruction = self.instructions[opcode]
        modes = f"{str(modes):>03}"[:-instruction.nparams-1:-1]
        params = list(range(i + 1, i + instruction.nparams + 1))

        instruction(*params, modes)
        return i + instruction.nparams + 1

    def execute(self, program, inputs=None, debug=False, step_limit=10000):
        self.program = program.copy()
        self.inputs = inputs
        instruction_ptr = 0
        step = 0
        while self.program[instruction_ptr] != IntcodeComputer.HALT:
            if debug:
                print("Instruction pointer", instruction_ptr)
                print(self.program)
                print()
            if step > step_limit:
                raise Exception("Step limit reached.")
            instruction_ptr = self.interp(instruction_ptr)
            step += 1
```

```python
computer = IntcodeComputer()
```

```python
computer.execute(program, iter([1]), debug=False)
```

<li style="border-color: rgb(204, 204, 204); color: rgb(204, 204, 204);">Opcode <code style="border: 1px solid rgb(51, 51, 64); color: rgb(204, 204, 204); background: rgb(16, 16, 26);">5</code> is <em style="border-color: rgb(255, 255, 255); color: rgb(255, 255, 255); text-shadow: rgb(255, 255, 255) 0px 0px 5px;">jump-if-true</em>: if the first parameter is <em style="border-color: rgb(255, 255, 255); color: rgb(255, 255, 255); text-shadow: rgb(255, 255, 255) 0px 0px 5px;">non-zero</em>, it sets the instruction pointer to the value from the second parameter. <p style="border-color: rgb(204, 204, 204); color: rgb(204, 204, 204);">Like all instructions, these instructions need to support <em style="border-color: rgb(255, 255, 255); color: rgb(255, 255, 255); text-shadow: rgb(255, 255, 255) 0px 0px 5px;">parameter modes</em> as described above.</p>
<p style="border-color: rgb(204, 204, 204); color: rgb(204, 204, 204);">Normally, after an instruction is finished, the instruction pointer increases by the number of values in that instruction. <em style="border-color: rgb(255, 255, 255); color: rgb(255, 255, 255); text-shadow: rgb(255, 255, 255) 0px 0px 5px;">However</em>, if the instruction modifies the instruction pointer, that value is used and the instruction pointer is <em style="border-color: rgb(255, 255, 255); color: rgb(255, 255, 255); text-shadow: rgb(255, 255, 255) 0px 0px 5px;">not automatically increased</em>.</p>
<p style="border-color: rgb(204, 204, 204); color: rgb(204, 204, 204);">For example, here are several programs that take one input, compare it to the value <code style="border: 1px solid rgb(51, 51, 64); color: rgb(204, 204, 204); background: rgb(16, 16, 26);">8</code>, and then produce one output:</p>
<ul style="border-color: rgb(204, 204, 204); color: rgb(204, 204, 204);">

```python
class IntcodeComputer:

    # Op codes
    ADD           = 1
    MUL           = 2
    READ          = 3
    WRITE         = 4
    JUMP_IF_TRUE  = 5
    JUMP_IF_FALSE = 6
    LESS_THAN     = 7
    EQUALS        = 8
    HALT          = 99

    def __init__(self, output_prefix=""):
        self.output_prefix = output_prefix
        self.instructions = {
            IntcodeComputer.ADD: self.add,
            IntcodeComputer.MUL: self.mul,
            IntcodeComputer.READ: self.read,
            IntcodeComputer.WRITE: self.write,
            IntcodeComputer.JUMP_IF_TRUE: self.jump_if_true,
            IntcodeComputer.JUMP_IF_FALSE: self.jump_if_false,
            IntcodeComputer.LESS_THAN: self.less_than,
            IntcodeComputer.EQUALS: self.equals
        }

    def __getitem__(self, address):
        return self.program[address]

    def __setitem__(self, address, value):
        self.program[address] = value

    def fetch(self, address, mode='1'):
        match mode:
            case '0': # position mode
                return self[self[address]]
            case '1': # immediate mode
                return self[address]
        raise Exception("Attempted to fetch with unknown mode:", mode)

    def store(self, address, value, mode='1'):
        match mode:
            case '0': # position mode
                self[self[address]] = value
            case '1': # immediate mode
                self[address] = value
            case _:
                raise Exception("Attempted to store with unknown mode:", mode)

    @static_vars(nparams=3)
    def add(self, a, b, dest, modes):
        self.store(self[dest], self.fetch(a, modes[0]) + self.fetch(b, modes[1]))
        self.instruction_pointer += 4

    @static_vars(nparams=3)
    def mul(self, a, b, dest, modes):
        self.store(self[dest], self.fetch(a, modes[0]) * self.fetch(b, modes[1]))
        self.instruction_pointer += 4

    @static_vars(nparams=1)
    def read(self, dest, _):
        self.store(self[dest], next(self.inputs))
        self.instruction_pointer += 2

    @static_vars(nparams=1)
    def write(self, address, modes):
        print(f"{self.output_prefix}{self.fetch(address, modes[0])}")
        self.instruction_pointer += 2

    @static_vars(nparams=2)
    def jump_if_true(self, condition, instruction_address, modes):
        if self.fetch(condition, modes[0]) == 0:
            self.instruction_pointer += 3
            return
        self.instruction_pointer = self.fetch(instruction_address, modes[1])

    @static_vars(nparams=2)
    def jump_if_false(self, condition, instruction_address, modes):
        if self.fetch(condition, modes[0]) != 0:
            self.instruction_pointer += 3
            return
        self.instruction_pointer = self.fetch(instruction_address, modes[1])

    @static_vars(nparams=3)
    def less_than(self, a, b, dest, modes):
        self.store(self[dest], int(self.fetch(a, modes[0]) < self.fetch(b, modes[1])))
        self.instruction_pointer += 4

    @static_vars(nparams=3)
    def equals(self, a, b, dest, modes):
        self.store(self[dest], int(self.fetch(a, modes[0]) == self.fetch(b, modes[1])))
        self.instruction_pointer += 4

    def step(self):
        modes, opcode = self[self.instruction_pointer] // 100, self[self.instruction_pointer] % 100
        instruction = self.instructions[opcode]
        modes = f"{str(modes):>03}"[:-instruction.nparams-1:-1]
        params = list(range(self.instruction_pointer + 1, self.instruction_pointer + instruction.nparams + 1))
        instruction(*params, modes)

    def execute(self, program, inputs=None, debug=False, step_limit=10000):
        self.program = program.copy()
        self.inputs = inputs
        self.instruction_pointer = 0
        step = 0
        while self.program[self.instruction_pointer] != IntcodeComputer.HALT:
            if debug:
                print("Instruction pointer", self.instruction_pointer)
                print(self.program)
                print()
            if step > step_limit:
                raise Exception("Step limit reached.")
            self.step()
            step += 1
```
