```python
d = graphviz.Digraph()
for row in circuit.rstrip().split('\n'):
    a, gate, b, _, c = row.split()
    node = " ".join([a, gate, b])
    d.edge(a, node)
    d.edge(b, node)
    d.edge(node, c)
d
```
