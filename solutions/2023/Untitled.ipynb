```python
# from aoc import *
from dataclasses import dataclass, field

node_map = {}

@dataclass
class Node:
    name: str
    links: set = field(default_factory=set)

f = open("./puzzle_inputs/25.txt")

g = graphviz.Graph()

for line in f:
    a, *others = re.findall("\w+", line)
    if a not in node_map:
        node_map[a] = Node(a)
    for other in others:
        if other not in node_map:
            node_map[other] = Node(other)
        node_map[other].links.add(a)
        g.edge(a, other)
        node_map[a].links.add(other)

print(len(node_map.keys()))

g
```
