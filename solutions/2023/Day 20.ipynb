```python
import graphviz
```

```python
with open("./puzzle_inputs/20.txt") as f:
    data = f.read()
```

```python
g = graphviz.Digraph("G")

module_types = {"rx": "", "broadcaster": ""}

for line in data.strip().split('\n'):
    source, _ = line.split(' -> ')
    if source[0] == "%":
        module_types[source[1:]] = "FF "
    elif source[0] == "&":
        module_types[source[1:]] = "Conj "

for line in data.strip().split('\n'):
    source, dests = line.split(' -> ')
    dests = dests.split(', ')
    for dest in dests:
        if source[0] in "%&":
            source = source[1:]
        g.edge(module_types[source]+source, module_types[dest]+dest)
g
```
