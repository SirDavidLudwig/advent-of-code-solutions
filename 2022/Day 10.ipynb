You pull it out of your pack, but the amount of water slowly draining from a big crack in its screen tells you it probably won't be of much immediate use.</p>
<p style="border-color: rgb(204, 204, 204); color: rgb(204, 204, 204);"><em style="border-color: rgb(255, 255, 255); color: rgb(255, 255, 255); text-shadow: rgb(255, 255, 255) 0px 0px 5px;">Unless</em>, that is, you can design a replacement for the device's video system! It seems to be some kind of <a href="https://en.wikipedia.org/wiki/Cathode-ray_tube" target="_blank" style="border-color: rgb(0, 153, 0); color: rgb(0, 153, 0);">cathode-ray tube</a> screen and simple CPU that are both driven by a precise <em style="border-color: rgb(255, 255, 255); color: rgb(255, 255, 255); text-shadow: rgb(255, 255, 255) 0px 0px 5px;">clock circuit</em>. The clock circuit ticks at a constant rate; each tick is called a <em style="border-color: rgb(255, 255, 255); color: rgb(255, 255, 255); text-shadow: rgb(255, 255, 255) 0px 0px 5px;">cycle</em>.</p>
<p style="border-color: rgb(204, 204, 204); color: rgb(204, 204, 204);">Start by figuring out the signal being sent by the CPU. The CPU has a single register, <code style="border: 1px solid rgb(51, 51, 64); color: rgb(204, 204, 204); background: rgb(16, 16, 26);">X</code>, which starts with the value <code style="border: 1px solid rgb(51, 51, 64); color: rgb(204, 204, 204); background: rgb(16, 16, 26);">1</code>. It supports only two instructions:</p>
<ul style="border-color: rgb(204, 204, 204); color: rgb(204, 204, 204);">
<li style="border-color: rgb(204, 204, 204); color: rgb(204, 204, 204);"><code style="border: 1px solid rgb(51, 51, 64); color: rgb(204, 204, 204); background: rgb(16, 16, 26);">addx V</code> takes <em style="border-color: rgb(255, 255, 255); color: rgb(255, 255, 255); text-shadow: rgb(255, 255, 255) 0px 0px 5px;">two cycles</em> to complete. <em style="border-color: rgb(255, 255, 255); color: rgb(255, 255, 255); text-shadow: rgb(255, 255, 255) 0px 0px 5px;">After</em> two cycles, the <code style="border: 1px solid rgb(51, 51, 64); color: rgb(204, 204, 204); background: rgb(16, 16, 26);">X</code> register is increased by the value <code style="border: 1px solid rgb(51, 51, 64); color: rgb(204, 204, 204); background: rgb(16, 16, 26);">V</code>. (<code style="border: 1px solid rgb(51, 51, 64); color: rgb(204, 204, 204); background: rgb(16, 16, 26);">V</code> can be negative.)</li>
<li style="border-color: rgb(204, 204, 204); color: rgb(204, 204, 204);"><code style="border: 1px solid rgb(51, 51, 64); color: rgb(204, 204, 204); background: rgb(16, 16, 26);">noop</code> takes <em style="border-color: rgb(255, 255, 255); color: rgb(255, 255, 255); text-shadow: rgb(255, 255, 255) 0px 0px 5px;">one cycle</em> to complete. It has no other effect.</li>
</ul>
<p style="border-color: rgb(204, 204, 204); color: rgb(204, 204, 204);">The CPU uses these instructions in a program (your puzzle input) to, somehow, tell the screen what to draw.</p>
<p style="border-color: rgb(204, 204, 204); color: rgb(204, 204, 204);">Consider the following small program:</p>
<pre style="border: 1px solid rgb(51, 51, 64); color: rgb(204, 204, 204); display: inline-block;"><code style="border-color: rgb(204, 204, 204); color: rgb(204, 204, 204);">noop

</code></pre>
<p style="border-color: rgb(204, 204, 204); color: rgb(204, 204, 204);">Execution of this program proceeds as follows:</p>
<ul style="border-color: rgb(204, 204, 204); color: rgb(204, 204, 204);">
<li style="border-color: rgb(204, 204, 204); color: rgb(204, 204, 204);">At the start of the first cycle, the <code style="border: 1px solid rgb(51, 51, 64); color: rgb(204, 204, 204); background: rgb(16, 16, 26);">noop</code> instruction begins execution. During the first cycle, <code style="border: 1px solid rgb(51, 51, 64); color: rgb(204, 204, 204); background: rgb(16, 16, 26);">X</code> is <code style="border: 1px solid rgb(51, 51, 64); color: rgb(204, 204, 204); background: rgb(16, 16, 26);">1</code>. After the first cycle, the <code style="border: 1px solid rgb(51, 51, 64); color: rgb(204, 204, 204); background: rgb(16, 16, 26);">noop</code> instruction finishes execution, doing nothing.</li>
<li style="border-color: rgb(204, 204, 204); color: rgb(204, 204, 204);">At the start of the second cycle, the <code style="border: 1px solid rgb(51, 51, 64); color: rgb(204, 204, 204); background: rgb(16, 16, 26);">addx 3</code> instruction begins execution. During the second cycle, <code style="border: 1px solid rgb(51, 51, 64); color: rgb(204, 204, 204); background: rgb(16, 16, 26);">X</code> is still <code style="border: 1px solid rgb(51, 51, 64); color: rgb(204, 204, 204); background: rgb(16, 16, 26);">1</code>.</li>
<li style="border-color: rgb(204, 204, 204); color: rgb(204, 204, 204);">During the third cycle, <code style="border: 1px solid rgb(51, 51, 64); color: rgb(204, 204, 204); background: rgb(16, 16, 26);">X</code> is still <code style="border: 1px solid rgb(51, 51, 64); color: rgb(204, 204, 204); background: rgb(16, 16, 26);">1</code>. After the third cycle, the <code style="border: 1px solid rgb(51, 51, 64); color: rgb(204, 204, 204); background: rgb(16, 16, 26);">addx 3</code> instruction finishes execution, setting <code style="border: 1px solid rgb(51, 51, 64); color: rgb(204, 204, 204); background: rgb(16, 16, 26);">X</code> to <code style="border: 1px solid rgb(51, 51, 64); color: rgb(204, 204, 204); background: rgb(16, 16, 26);">4</code>.</li>
<li style="border-color: rgb(204, 204, 204); color: rgb(204, 204, 204);">At the start of the fourth cycle, the <code style="border: 1px solid rgb(51, 51, 64); color: rgb(204, 204, 204); background: rgb(16, 16, 26);">addx -5</code> instruction begins execution.
```python
import numpy as np
```

```python
with open("./puzzle_inputs/10.txt") as f:
    ops = [op.strip() for op in f]
len(ops)
```

```python
num_cycles = 0
register = 1
signals = []
for op in ops:
    num_cycles += 1
    if (num_cycles - 20) % 40 == 0:
        signals.append(num_cycles*register)
    if op == "noop":
        continue

    num_cycles += 1
    if (num_cycles - 20) % 40 == 0:
        signals.append(num_cycles*register)
    register += int(op[4:])
np.sum(signals)
```
