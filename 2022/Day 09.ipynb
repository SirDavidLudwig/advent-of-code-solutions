<h2>--- Day 9: Rope Bridge ---</h2><p>This rope bridge creaks as you walk along it. You aren't sure how old it is, or whether it can even support your weight.</p>
<p>It seems to support the Elves just fine, though. The bridge spans a gorge which was carved out by the massive river far below you.</p>
<p>You step carefully; as you do, the ropes stretch and twist. You decide to distract yourself by modeling rope physics; maybe you can even figure out where <em style="color: #fff; text-shadow: 0 0 5px #fff;">not</em> to step.</p>
<p>Consider a rope with a knot at each end; these knots mark the <em style="color: #fff; text-shadow: 0 0 5px #fff;">head</em> and the <em style="color: #fff; text-shadow: 0 0 5px #fff;">tail</em> of the rope. If the head moves far enough away from the tail, the tail is pulled toward the head.</p>
<p>Due to nebulous reasoning involving <a href="https://en.wikipedia.org/wiki/Planck_units#Planck_length" target="_blank">Planck lengths</a>, you should be able to model the positions of the knots on a two-dimensional grid. Then, by following a hypothetical <em style="color: #fff; text-shadow: 0 0 5px #fff;">series of motions</em> (your puzzle input) for the head, you can determine how the tail will move.</p>
<p><span title="I'm an engineer, not a physicist!">Due to the aforementioned Planck lengths</span>, the rope must be quite short; in fact, the head (<code>H</code>) and tail (<code>T</code>) must <em style="color: #fff; text-shadow: 0 0 5px #fff;">always be touching</em> (diagonally adjacent and even overlapping both count as touching):</p>

<p>After simulating the rope, you can count up all of the positions the <em style="color: #fff; text-shadow: 0 0 5px #fff;">tail visited at least once</em>. In this diagram, <code>s</code> again marks the starting position (which the tail also visited) and <code>#</code> marks other positions the tail visited:</p>

<p>So, there are <code><em style="color: #fff; text-shadow: 0 0 5px #fff;">13</em></code> positions the tail visited at least once.</p>
<p>Simulate your complete hypothetical series of motions. <em style="color: #fff; text-shadow: 0 0 5px #fff;">How many positions does the tail of the rope visit at least once?</em></p>

```python
import numpy as np
```

```python
with open("./puzzle_inputs/09.txt") as f:
    moves = [(move[0], int(move[1:])) for move in f]
len(moves)
```

```python
move_map = {
    'U': np.array([-1,  0]),
    'D': np.array([ 1,  0]),
    'L': np.array([ 0, -1]),
    'R': np.array([ 0,  1])
}
```

```python
tail_positions = set([(0, 0)])
head, tail = np.zeros((2, 2))

for direction, units in moves:
    delta = move_map[direction]
    for _ in range(units):
        head += delta
        if np.linalg.norm(head - tail) >= 2.0:
            tail = head - delta
            tail_positions.add(tuple(tail))
len(tail_positions)
```

```python
tail_positions = set([(0, 0)])
knots = np.zeros((10, 2), dtype=int)

for direction, units in moves:
    delta = move_map[direction]
    for j in range(units):
        knots[0] += move_map[direction]
        for i in range(1, len(knots)):
            if np.linalg.norm(knots[i-1] - knots[i]) < 2:
                break
            knots[i] += np.clip(knots[i-1] - knots[i], -1, 1)
        tail_positions.add(tuple(knots[-1]))
len(tail_positions)
```
